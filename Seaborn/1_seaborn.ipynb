# KDE PLOT -- KERNAL DENSITY ESTIMATION
is a smooth curve that shows the distribution of a variable (like a smoothed histogram).
Total area of any frequency distri is always 1 or 100%, i.e sum of every (distri/total)

```python
# stat can be 'count', 'frequency', 'density', 'probability'
sns.histplot(data=df, x='MathScore',  stat='density', multiple='stack', hue='Gender', fill=True, alpha=0.3, kde=True)
plt.grid(True)
# density normalizes the area under the histogram to 1
# density curve helps us visualize the shape of the distribution of data
```

# RUG PLOT
A rug plot shows each data point as a tiny tick mark on an axis. Its great for seeing where values are concentrated, usually alongside a KDE or histogram.

```python
plt.Figure(figsize=(15, 5))

sns.rugplot(data=df, x='MathScore', height=0.2)
```

```python
sns.kdeplot(data=df, x='MathScore', y='EnglishScore', hue='Gender', fill=True)
sns.rugplot(data=df, x='MathScore', y='EnglishScore', hue='Gender', height=-0.04, clip_on=False)
```

```python
sns.scatterplot(data=df, x='MathScore', y='EnglishScore', hue='Gender')
sns.rugplot(data=df, x='MathScore', y='EnglishScore', hue='Gender', height=-0.04, clip_on=False)
```

# ECDF -- Empirical Cumulative Distribustion Function
is a plot that shows, for every value x, the proportion of observations  x. So if ECDF at 50 is 0.70, it means 70% of your data is 50 or below.

```python
sns.ecdfplot(data=df, x='MathScore', hue='Gender', stat='count')

plt.grid(True, alpha=0.3)
```

# DIS PLOT
a displot (sns.displot) is a distribution plot function. It can create histograms, KDE plots, ECDF plots, and it can also split the plots into panels (facets) by category.

```python
sns.displot(data=df, x='MathScore', kde=True)
```

```python
sns.displot(data=df, x='MathScore', kde=True, rug=True, rug_kws={'height':0.05}, hue='Gender', multiple='stack', col='Gender')
# sns.displot(data=df, x='MathScore', kde=True, rug=True, rug_kws={'height':0.05}, hue='Gender', multiple='stack', col='Gender', row='Country')
```

```python
sns.displot(data=df, x='MathScore', kind='ecdf', hue='Gender', col='Gender')
```

# jointplot 
jointplot in Seaborn shows the relationship between two numeric variables plus their individual distributions (hist/KDE) on the top and right margins.
 scatter plot in the middle + histograms/KDE around it.

```python
# Joint plot is the multiple of plot

sns.jointplot(data=df, x='MathScore', y='EnglishScore', height=5)
plt.grid(True, alpha=0.5)
```

```python
sns.jointplot(data=df, x='MathScore', y='EnglishScore',
               kind='scatter', 
               hue='Gender', 
               palette='BuPu')
# by default it shows the scatter, with the kind= , it can be changed
```

```python
sns.jointplot(data=df, x='MathScore', y='EnglishScore',
              joint_kws=dict(marker='*', color='red'), 
              marginal_kws=dict(color='green', kde=True), 
              height=7, 
              ratio=3, 
              space=2, 
              marginal_ticks=True)
```

```python
g = sns.jointplot(data=df, x='MathScore', y='EnglishScore', height=6, color='r', joint_kws=dict(color='green'), marginal_kws=dict(color='black'))
g.plot_joint(sns.kdeplot, color='red')
g.plot_joint(sns.rugplot)
```


# pair plot
A pair plot (sns.pairplot) creates a grid of plots for multiple numeric columns:
diagonal = each columns distribution (hist/KDE)
off-diagonal = scatter plots showing relationships between every pair
Its great for quick EDA, but for big datasets you should sample. 
it plot pairwise relationship in a data set 

```python
# Pairplot-- it plot pairwise relationship in a data set 

sns.pairplot(df)
```

```python
sns.pairplot(data=df,x_vars=['Age', 'MathScore', 'EnglishScore'], y_vars=['Age', 'MathScore', 'EnglishScore'], diag_kind='kde', hue='Gender', diag_kws=dict(color='r', fill=False))
```

Corner=True, enable us to see the lower triangle after the diagonal

```python
sns.pairplot(data=df, x_vars=['Age', 'MathScore', 'EnglishScore'], y_vars=['Age', 'MathScore', 'EnglishScore'], corner=True,  diag_kws=dict(kde=True),  plot_kws=dict(color='green'))
```

```python
g = sns.pairplot(data=df, x_vars=['Age', 'MathScore', 'EnglishScore'], y_vars=['Age', 'MathScore', 'EnglishScore'], plot_kws=dict(color='r'), diag_kws={'kde':True})
g.map_lower(sns.kdeplot)
```

# RELATIONAL PLTS
Relational plots in Seaborn are mainly done with sns.relplot(). Its a figure-level function (like catplot and displot) and it can create scatter or line plots, plus it can split into panels using row= and col=.

# SCATTER PLOT
Its a diagram which shows the relationship between two variables by plotting the point

```python
# sns.scatterplot for a single plot with simple control.
# Use sns.relplot when you want facets (row, col) or a quick multi-plot layout

df_small = df.sample(50, random_state=42)  # random 50 samples

sns.scatterplot(data=df_small, x='AttendancePct', y='MathScore', s=200, edgecolor='black',
                # hue='Gender',
                hue='ClassLevel',
                style='Gender',
                markers={'MALE': 'o', 'FEMALE': 'v'}
                )               # this invole two categorical variable.
plt.grid(True)
```

```python
sns.scatterplot(data=df_small, x='AttendancePct', y='MathScore', 
                hue='Gender',
                size="Gender",
                sizes=(100, 200),
                legend='full',
                # hue='ClassLevel',
                palette='Set2_r',
                )              # hue and size -- two categorical relationship
```

# LINE PLOT
shows the relationship between  variable, majorly used in time series analysis 

Bootstrapping = estimate uncertainty by repeatedly resampling your data (with replacement) and recomputing the statistic.

```python
df_small.head()
```

```python
sns.lineplot(data=df_small, x='Age', y='MathScore',
             errorbar=None,
                estimator=sum,
                # n_boot=1000,
                hue='Gender',
                )


# ci= or errorbar=None === confident interval
# by default it estimate the mean of each AGE and ci is 95% by default
# from each point of the scattered point on a particular x axis, the sum of the point are collected on the lineplot either by 
plt.grid(True)
```

```python
sns.lineplot(data=df_small, x='Age', y='MathScore',
               errorbar=None,
               estimator=sum,
               hue='Gender',
               # style='Gender',
               # markers=True,
               # dashes=False,
               size='Gender',
               sizes=(1, 3),
               )

# size, hue and style can be use to categorize the plot
```
